```python
#Removing the pattern 
import re

def month_detection(value):
    month =['January', 'February','March','April','May','June', 'July','August','Septemeber','October','November','Decemeber']
    month_short = ['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']
    value = str(value).strip()
    if re.fullmatch(r'[A-Za-z{3,}]',value):
        g1 = re.fullmatch(r'[A-Za-z{3,}]',value)
        if (g1 in month) | (g1 in month_short):
            return g1
        else:
            return -1
    elif re.fullmatch(r'\d{4}',value): #if there is a regex code, then print no month deteted)
        #print('no month detected')
        return -1  #return -1 as no month detected
    elif re.fullmatch(r'[\"A-Za-z\.\(\)\?&]*', value):
        return -1
    elif re.fullmatch(r'\d{3,5}',value):
        return -1
    elif re.fullmatch(r'\d{2,4}s', value):
        return -1
    elif re.fullmatch(r'\d{4}/d{4}\)?', value):
        return -1
    elif re.fullmatch(r'\d{4}\.\d{2}\.',value):
        return -1
    elif re.fullmatch(r'([A-Za-z]+) \d+', value):
        g1=re.fullmatch(r'([A-Za-z]+) \d+',value).group(1)  # group 0 is the full match
        if (g1 in month) | (g1 in month_short):
            return g1
        else:
            return -1
    elif re.fullmatch(r'\d{1,2}-(\w{2,3})-\d{2,4}',value):
        return re.fullmatch(r'\d{1,2}-(\w{2,3})-\d{2,4}',value).group(1)
    #elif re.fullmatch(r'(\w{2,3})-\d{2,4}',value):
        #return re.fullmatch(r'(\w{2,3})-\d{2,4}',value).group(1)
    elif re.fullmatch(r'Reported \d{1,2}-(\w{2,3})-\d{2,4}',value):
        return re.fullmatch(r'Reported \d{1,2}-(\w{2,3})-\d{2,4}',value).group(1)
    elif re.fullmatch(r'([A-Za-z]{3})-\d{2,4}',value):
        return re.fullmatch(r'([A-Za-z]{3})-\d{2,4}',value).group(1)
    else:
        return -1 #delete all the non defined value
    
    
    
```
